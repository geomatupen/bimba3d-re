# Multiplier Prediction Test
Test two independently implemented prediction modes against project EXIF features:
1. **Featurewise Ridge** — scans log-space candidate actions, selects action that maximises predicted reward `f(x,a) → reward`
2. **Featurewise MLP** — single forward pass directly predicts multipliers `f(x) → multiplier`

Each section is **independent** — Ridge uses only `train_offline_models.py` functions, MLP uses only PyTorch. Set your paths in Cell 2, then run either section.

```python
# ── CONFIGURATION ─────────────────────────────────────────────────────────────
# Set these paths before running any section below.
import sys
from pathlib import Path

WORKSPACE_ROOT = Path(r"D:\bimba3d-re")

# Pipeline folder that contains project subdirectories.
# Each project folder must have an exif_features.json file inside it.
PIPELINE_FOLDER = Path(r"E:\Thesis\PipelineProjects\Training_June_2")

# Offline trained model folder (contains quality_model_*.json, convergence_model_*.json)
# Leave as None to auto-detect from pipeline folder.
OFFLINE_MODEL_DIR = None

# For MLP: shared_models directory (contains neural_contextual/ subfolder with featurewise*.pt)
# Leave as None to auto-detect from pipeline folder.
SHARED_MODELS_DIR = None

# Ridge prediction objective: "quality" or "convergence"
OBJECTIVE = "quality"

# Number of log-space candidate points for ridge bounded search (more = finer grid)
CANDIDATE_POINTS = 21

# ── add workspace to Python path ──────────────────────────────────────────────
if str(WORKSPACE_ROOT) not in sys.path:
    sys.path.insert(0, str(WORKSPACE_ROOT))

# Auto-detect dirs if not set
if OFFLINE_MODEL_DIR is None:
    OFFLINE_MODEL_DIR = PIPELINE_FOLDER / "_offline_training" / "models"
if SHARED_MODELS_DIR is None:
    SHARED_MODELS_DIR = PIPELINE_FOLDER / "shared_models"

print(f"Pipeline folder   : {PIPELINE_FOLDER}")
print(f"Offline model dir : {OFFLINE_MODEL_DIR}  exists={OFFLINE_MODEL_DIR.exists()}")
print(f"Shared models dir : {SHARED_MODELS_DIR}  exists={SHARED_MODELS_DIR.exists()}")
```

```python
# ── COMMON: Load EXIF features for all projects ────────────────────────────────
import json, math
import numpy as np

def load_all_project_exif(pipeline_folder: Path) -> dict:
    """Scan pipeline folder for exif_features.json in each project subdirectory."""
    projects = {}
    for project_dir in sorted(pipeline_folder.iterdir()):
        if not project_dir.is_dir():
            continue
        if project_dir.name.startswith("_") or project_dir.name in {"shared_models", "training_pipelines"}:
            continue
        exif_file = project_dir / "exif_features.json"
        if exif_file.exists():
            try:
                payload = json.loads(exif_file.read_text(encoding="utf-8"))
                # exif_features.json is either {"features": {...}, ...} or a raw feature dict
                features = payload.get("features") if isinstance(payload.get("features"), dict) else payload
                if isinstance(features, dict) and features:
                    projects[project_dir.name] = {
                        "features": features,
                        "project_dir": project_dir,
                        "mode": payload.get("mode", "?"),
                        "heuristic_preset": payload.get("heuristic_preset", "?"),
                        "extracted_at": payload.get("extracted_at", "?"),
                    }
            except Exception as e:
                print(f"  [warn] {project_dir.name}: {e}")
    return projects

projects = load_all_project_exif(PIPELINE_FOLDER)
print(f"Found {len(projects)} projects with EXIF features:\n")
for name, info in projects.items():
    print(f"  {name:45s}  mode={info['mode']:35s}  features={len(info['features'])}")
```

## Section 1 — Featurewise Ridge Regression

**How it works:**
- Loads offline-trained ridge model (`quality_model_*.json` or `convergence_model_*.json`)
- For each project, builds a per-group z-score feature vector from EXIF data
- Scans `CANDIDATE_POINTS` candidate actions in log-space within group bounds
- Selects the action with the highest predicted reward `φ(x, a) @ θ`
- Falls back to **neutral (1.0)** when the score surface is flat (model has no signal)

> Design vector: `φ(x, a) = [x, a, a², x_no_intercept·a]`

```python
# ── RIDGE: Load offline model ──────────────────────────────────────────────────
from bimba3d_backend.scripts.train_offline_models import (
    FEATUREWISE_GROUP_FEATURES,
    GROUP_KEYS,
    GROUP_BOUNDS,
    GROUP_MULTIPLIERS_MAP,
    apply_scaler_to_vector,
    _build_reward_design_vector,
)

def load_ridge_model(offline_model_dir: Path, objective: str) -> dict | None:
    """Load the newest quality or convergence model JSON."""
    prefix = "quality_model" if objective == "quality" else "convergence_model"
    versioned = sorted(offline_model_dir.glob(f"{prefix}_*.json"), key=lambda p: p.stem, reverse=True)
    path = versioned[0] if versioned else offline_model_dir / f"{prefix}.json"
    if not path.exists():
        print(f"  [ERROR] No {prefix} model found in {offline_model_dir}")
        return None
    payload = json.loads(path.read_text(encoding="utf-8"))
    # Unwrap offline_model_v3 schema wrapper
    inner = payload.get("model", payload) if payload.get("schema") == "offline_model_v3" else payload
    if "model_family" not in inner:
        inner["model_family"] = payload.get("model_family", "ridge_reward_optimizer")
    print(f"  Loaded : {path.name}")
    print(f"  runs   : {inner.get('runs', 0)}")
    print(f"  reward_mean : {inner.get('reward_mean', 0):.4f}")
    print(f"  lambda_ridge: {inner.get('lambda_ridge', '?')}")
    print(f"  candidates  : {inner.get('candidate_points', '?')}")
    return inner

ridge_model = load_ridge_model(OFFLINE_MODEL_DIR, OBJECTIVE)
print(f"\nmodel_family = {ridge_model.get('model_family') if ridge_model else 'N/A'}")
```

```python
# ── RIDGE: Self-contained prediction logic (no MLP, no mixing) ────────────────

SIGNAL_THRESHOLD = 1e-6

RIDGE_PARAMETER_GROUPS = {
    "geometry_lr_mult":   ["position_lr_init_mult", "scaling_lr_mult", "rotation_lr_mult"],
    "appearance_lr_mult": ["feature_lr_mult", "opacity_lr_mult", "lambda_dssim_mult"],
    "densification_mult": ["densify_grad_threshold_mult", "opacity_threshold_mult"],
}

def ridge_select_group(model: dict, group_key: str, x_features: dict, n_points: int = 21) -> dict:
    """
    Ridge reward-optimizer inference for one parameter group.
    Scans n_points log-space candidate actions in [lo, hi] and picks the
    action with the highest predicted reward  φ(x,a) @ θ.
    Returns neutral (1.0) when the score surface is flat.
    """
    md = model["models"][group_key]
    A = np.array(md["A"], dtype=np.float64)
    b = np.array(md["b"], dtype=np.float64)
    n_obs = int(md.get("n", 0))

    try:
        theta = np.linalg.solve(A, b)
    except np.linalg.LinAlgError:
        theta = np.linalg.pinv(A) @ b

    scaler = model.get("feature_scalers", {}).get(group_key, {})
    x = apply_scaler_to_vector(x_features, group_key, scaler)

    lo_mult, hi_mult = GROUP_BOUNDS[group_key]
    lo, hi = math.log(lo_mult), math.log(hi_mult)
    candidates = np.linspace(lo, hi, n_points, dtype=np.float64)

    # Score every candidate action
    scores = [float(_build_reward_design_vector(x, float(a)) @ theta) for a in candidates]
    score_spread = max(scores) - min(scores)

    if score_spread < SIGNAL_THRESHOLD:
        # Model has no discriminating signal → use neutral multiplier (1.0)
        best_log = 0.0
        has_signal = False
    else:
        best_log = float(candidates[int(np.argmax(scores))])
        has_signal = True

    best_mult = float(np.clip(math.exp(best_log), lo_mult, hi_mult))

    return {
        "multiplier":    best_mult,
        "log_multiplier": math.log(best_mult),
        "score_spread":  score_spread,
        "has_signal":    has_signal,
        "n_obs":         n_obs,
        "theta_norm":    float(np.linalg.norm(theta)),
        "scores":        scores,
        "candidates":    candidates.tolist(),
    }


def ridge_predict_project(model: dict, x_features: dict, n_points: int = 21) -> dict:
    """Run ridge inference for all 3 groups, expand to 8 individual multipliers."""
    group_results = {g: ridge_select_group(model, g, x_features, n_points) for g in GROUP_KEYS}
    multipliers = {}
    for group_key, member_keys in RIDGE_PARAMETER_GROUPS.items():
        m = group_results[group_key]["multiplier"]
        for key in member_keys:
            multipliers[key] = m
    return {
        "multipliers":    multipliers,
        "group_results":  group_results,
        "has_signal":     all(group_results[g]["has_signal"] for g in GROUP_KEYS),
    }

print("Ridge prediction logic loaded.")
```

```python
# ── RIDGE: Run predictions and print table ─────────────────────────────────────
import pandas as pd

if ridge_model is None:
    print("No ridge model loaded. Train an offline model first from the training pipeline.")
else:
    n_pts = int(ridge_model.get("candidate_points") or CANDIDATE_POINTS)
    rows = []
    for project_name, info in projects.items():
        result = ridge_predict_project(ridge_model, info["features"], n_pts)
        m = result["multipliers"]
        g = result["group_results"]
        rows.append({
            "project":       project_name,
            "signal":        "✓" if result["has_signal"] else "⚠ no signal",
            # 3 group multipliers
            "geo_mult":      round(g["geometry_lr_mult"]["multiplier"], 4),
            "app_mult":      round(g["appearance_lr_mult"]["multiplier"], 4),
            "den_mult":      round(g["densification_mult"]["multiplier"], 4),
            # 8 expanded individual multipliers
            "pos_lr":        round(m["position_lr_init_mult"], 4),
            "scl_lr":        round(m["scaling_lr_mult"], 4),
            "rot_lr":        round(m["rotation_lr_mult"], 4),
            "fea_lr":        round(m["feature_lr_mult"], 4),
            "opa_lr":        round(m["opacity_lr_mult"], 4),
            "lam_dssim":     round(m["lambda_dssim_mult"], 4),
            "den_grad":      round(m["densify_grad_threshold_mult"], 4),
            "opa_thr":       round(m["opacity_threshold_mult"], 4),
            # Diagnostics
            "n_obs":         g["geometry_lr_mult"]["n_obs"],
            "spr_geo":       round(g["geometry_lr_mult"]["score_spread"], 5),
            "spr_app":       round(g["appearance_lr_mult"]["score_spread"], 5),
            "spr_den":       round(g["densification_mult"]["score_spread"], 5),
        })

    df_ridge = pd.DataFrame(rows)
    pd.set_option("display.max_columns", None)
    pd.set_option("display.width", 240)
    pd.set_option("display.float_format", "{:.4f}".format)

    print(f"=== RIDGE PREDICTION  (objective={OBJECTIVE}, candidates={n_pts}) ===\n")
    print(df_ridge.to_string(index=False))

    no_signal = df_ridge[df_ridge["signal"] != "✓"]
    if not no_signal.empty:
        print(f"\n⚠  {len(no_signal)} project(s) got neutral multipliers (1.0) — model has no signal:")
        for _, r in no_signal.iterrows():
            print(f"   {r['project']}  (spread_geo={r['spr_geo']}, spread_app={r['spr_app']}, spread_den={r['spr_den']})")
        print("   → Rebuild offline dataset and retrain offline model with more diverse run data.")
```

```python
# ── RIDGE: Score surface plot for one project ─────────────────────────────────
# Shows the predicted reward curve across the candidate grid for each group.
# A flat line = no signal.  A clear peak = the model has learned something.
import matplotlib.pyplot as plt

PLOT_PROJECT = list(projects.keys())[0]   # change to any project name from the table

if ridge_model is not None:
    x_features = projects[PLOT_PROJECT]["features"]
    n_pts = int(ridge_model.get("candidate_points") or CANDIDATE_POINTS)

    fig, axes = plt.subplots(1, 3, figsize=(16, 4))
    fig.suptitle(f"Ridge reward surface  —  {PLOT_PROJECT}  (objective={OBJECTIVE})", fontsize=12)

    for ax, group_key in zip(axes, GROUP_KEYS):
        r = ridge_select_group(ridge_model, group_key, x_features, n_pts)
        mults = [math.exp(a) for a in r["candidates"]]
        ax.plot(mults, r["scores"], marker="o", markersize=3, linewidth=1.5)
        ax.axvline(r["multiplier"], color="red", linestyle="--", linewidth=1.5,
                   label=f"selected={r['multiplier']:.3f}")
        ax.axvline(1.0, color="gray", linestyle=":", linewidth=1, label="neutral")
        title = group_key.replace("_mult", "")
        ax.set_title(f"{title}\n" + ("✓ signal" if r["has_signal"] else "⚠ flat / no signal"),
                     fontsize=9)
        ax.set_xlabel(f"multiplier  (spread={r['score_spread']:.2e}, n={r['n_obs']})", fontsize=8)
        ax.set_ylabel("predicted reward")
        ax.legend(fontsize=8)

    plt.tight_layout()
    plt.show()
```

## Section 2 — Featurewise Neural MLP

**How it works:**
- Loads the trained featurewise MLP (`featurewise_*.pt` from `shared_models/neural_contextual/`)
- Three separate tiny networks, one per group, each with its own feature subset
- Single forward pass per project — no reward scoring, no grid search
- Outputs raw multiplier for each group, clamped to bounds

> Architecture per group: `input(d) → Dense(8, ReLU, Dropout) → Dense(4, ReLU, Dropout) → Dense(1)`

```python
# ── MLP: Check PyTorch and load checkpoint ────────────────────────────────────
try:
    import torch
    import torch.nn as nn
    HAS_TORCH = True
    print(f"PyTorch {torch.__version__}")
except ImportError:
    HAS_TORCH = False
    print("[ERROR] PyTorch not installed. Run:  pip install torch")

def load_featurewise_checkpoint(shared_models_dir: Path):
    """Load newest featurewise MLP .pt file."""
    if not HAS_TORCH:
        return None, None
    neural_dir = shared_models_dir / "neural_contextual"
    versioned = sorted(neural_dir.glob("featurewise_*.pt"), key=lambda p: p.stem, reverse=True)
    if not versioned:
        print(f"  [ERROR] No featurewise*.pt found in {neural_dir}")
        return None, None
    path = versioned[0]
    ckpt = torch.load(path, map_location="cpu", weights_only=False)
    print(f"  Loaded : {path.name}")
    print(f"  geo_dim={ckpt['geo_dim']}  app_dim={ckpt['app_dim']}  den_dim={ckpt['den_dim']}")
    print(f"  hidden={ckpt.get('hidden',8)}  dropout={ckpt.get('dropout',0.2)}  objective={ckpt.get('objective','?')}")
    # Load companion metadata if present
    meta_path = path.with_name(path.stem + "_metadata.json")
    meta = json.loads(meta_path.read_text()) if meta_path.exists() else {}
    if meta:
        print(f"  training_samples={meta.get('training_samples','?')}  "
              f"epochs={meta.get('epochs_trained','?')}  "
              f"best_val_loss={meta.get('best_val_loss',float('nan')):.6f}")
    return ckpt, meta

mlp_checkpoint, mlp_meta = load_featurewise_checkpoint(SHARED_MODELS_DIR)
```

```python
# ── MLP: Architecture (self-contained, no imports from neural_contextual_learner) ──

if HAS_TORCH:
    class _GroupMLP(nn.Module):
        """Tiny single-group MLP: input(d) → hidden → hidden//2 → 1 output."""
        def __init__(self, input_dim, hidden=8, dropout=0.2):
            super().__init__()
            self.net = nn.Sequential(
                nn.Linear(input_dim, hidden), nn.ReLU(), nn.Dropout(dropout),
                nn.Linear(hidden, max(1, hidden // 2)), nn.ReLU(), nn.Dropout(dropout),
                nn.Linear(max(1, hidden // 2), 1),
            )
        def forward(self, x):
            return self.net(x)

    class _FeaturewiseMLP(nn.Module):
        """Three separate tiny MLPs, one per group."""
        def __init__(self, geo_dim, app_dim, den_dim, hidden=8, dropout=0.2):
            super().__init__()
            self.geometry_net      = _GroupMLP(geo_dim, hidden, dropout)
            self.appearance_net    = _GroupMLP(app_dim, hidden, dropout)
            self.densification_net = _GroupMLP(den_dim, hidden, dropout)
        def forward(self, x_geo, x_app, x_den):
            return torch.cat([
                self.geometry_net(x_geo),
                self.appearance_net(x_app),
                self.densification_net(x_den),
            ], dim=1)   # (batch, 3)

    print("MLP architecture defined.")
```

```python
# ── MLP: Feature vector builder (same as training — must match exactly) ────────
# Copied from contextual_continuous_learner._build_featurewise_vector so this
# section has zero external dependency on the production code.

MLP_GROUP_FEATURES = {
    "geometry_lr_mult": [
        "intercept", "focal_norm", "gsd_norm", "overlap_proxy", "coverage_spread",
        "camera_angle_bucket", "heading_consistency", "blur_motion_risk",
        "terrain_roughness_proxy", "vegetation_cover",
    ],
    "appearance_lr_mult": [
        "intercept", "iso_norm", "image_resolution_norm", "blur_motion_risk",
        "texture_density", "vegetation_cover", "vegetation_complexity",
    ],
    "densification_mult": [
        "intercept", "gsd_norm", "overlap_proxy", "coverage_spread",
        "camera_angle_bucket", "texture_density", "blur_motion_risk",
        "terrain_roughness_proxy", "vegetation_complexity",
    ],
}

MLP_GROUP_BOUNDS = {
    "geometry_lr_mult":   (0.5, 1.5),
    "appearance_lr_mult": (0.5, 1.5),
    "densification_mult": (0.7, 1.3),
}

MLP_PARAMETER_GROUPS = {
    "geometry_lr_mult":   ["position_lr_init_mult", "scaling_lr_mult", "rotation_lr_mult"],
    "appearance_lr_mult": ["feature_lr_mult", "opacity_lr_mult", "lambda_dssim_mult"],
    "densification_mult": ["densify_grad_threshold_mult", "opacity_threshold_mult"],
}

def _build_group_feature_vector(features: dict, group_key: str) -> np.ndarray:
    """Build the per-group feature vector using the same mapping as training."""
    names = MLP_GROUP_FEATURES[group_key]
    x = []
    for name in names:
        if name == "intercept":
            x.append(1.0)
        elif name == "focal_norm":
            x.append(math.log(max(float(features.get("focal_length_mm", 24.0) or 24.0), 1e-9)))
        elif name == "iso_norm":
            x.append(math.log(max(float(features.get("iso", 400.0) or 400.0), 1e-9)))
        elif name == "image_resolution_norm":
            w = float(features.get("img_width_median", 4000.0) or 4000.0)
            h = float(features.get("img_height_median", 3000.0) or 3000.0)
            x.append((w * h) / 1e6)
        elif name == "gsd_norm":
            x.append(math.log(max(float(features.get("gsd_median", 0.05) or 0.05), 1e-9)))
        elif name == "overlap_proxy":
            x.append(float(features.get("overlap_proxy", 0.5) or 0.5))
        elif name == "coverage_spread":
            x.append(float(features.get("coverage_spread", 0.5) or 0.5))
        elif name == "camera_angle_bucket":
            x.append(float(features.get("camera_angle_bucket", 0) or 0))
        elif name == "heading_consistency":
            x.append(float(features.get("heading_consistency", 0.5) or 0.5))
        elif name == "texture_density":
            x.append(float(features.get("texture_density", 0.5) or 0.5))
        elif name == "blur_motion_risk":
            x.append(float(features.get("blur_motion_risk", 0.5) or 0.5))
        elif name == "terrain_roughness_proxy":
            v = features.get("terrain_roughness_proxy", features.get("terrain_roughness", 0.5))
            x.append(float(v or 0.5))
        elif name == "vegetation_cover":
            v = features.get("vegetation_cover_percentage", features.get("vegetation_cover", 0.5))
            x.append(float(v or 0.5))
        elif name == "vegetation_complexity":
            v = features.get("vegetation_complexity_score", features.get("vegetation_complexity", 0.5))
            x.append(float(v or 0.5))
        else:
            x.append(0.0)
    return np.array(x, dtype=np.float32)

print("MLP feature vector builder defined.")
```

```python
# ── MLP: Self-contained prediction logic (no ridge, no mixing) ────────────────

def mlp_predict_project(ckpt: dict, x_features: dict) -> dict:
    """
    Featurewise MLP inference:
    1. Build per-group feature vectors from EXIF dict
    2. Run each group through its dedicated tiny network (greedy, dropout off)
    3. Clamp raw output to group bounds
    4. Expand 3 group multipliers to 8 individual multipliers
    """
    model = _FeaturewiseMLP(
        geo_dim=ckpt["geo_dim"],
        app_dim=ckpt["app_dim"],
        den_dim=ckpt["den_dim"],
        hidden=ckpt.get("hidden", 8),
        dropout=ckpt.get("dropout", 0.2),
    )
    model.load_state_dict(ckpt["state_dict"])
    model.eval()

    x_geo = torch.tensor(_build_group_feature_vector(x_features, "geometry_lr_mult")).unsqueeze(0)
    x_app = torch.tensor(_build_group_feature_vector(x_features, "appearance_lr_mult")).unsqueeze(0)
    x_den = torch.tensor(_build_group_feature_vector(x_features, "densification_mult")).unsqueeze(0)

    with torch.no_grad():
        raw = model(x_geo, x_app, x_den).squeeze(0).numpy()

    # Clamp to bounds
    def _clamp(v, lo, hi): return float(max(lo, min(hi, v)))
    geo_mult = _clamp(raw[0], *MLP_GROUP_BOUNDS["geometry_lr_mult"])
    app_mult = _clamp(raw[1], *MLP_GROUP_BOUNDS["appearance_lr_mult"])
    den_mult = _clamp(raw[2], *MLP_GROUP_BOUNDS["densification_mult"])

    group_multipliers = {
        "geometry_lr_mult":   geo_mult,
        "appearance_lr_mult": app_mult,
        "densification_mult": den_mult,
    }

    # Expand 3 group → 8 individual multipliers
    multipliers = {}
    for group_key, member_keys in MLP_PARAMETER_GROUPS.items():
        for key in member_keys:
            multipliers[key] = group_multipliers[group_key]

    clamped = [geo_mult != float(raw[0]), app_mult != float(raw[1]), den_mult != float(raw[2])]

    return {
        "multipliers":       multipliers,
        "group_multipliers": group_multipliers,
        "raw_output":        raw.tolist(),
        "any_clamped":       any(clamped),
    }

print("MLP prediction logic loaded.")
```

```python
# ── MLP: Run predictions and print table ───────────────────────────────────────
if not HAS_TORCH:
    print("PyTorch not available.")
elif mlp_checkpoint is None:
    print("No MLP model loaded. Train a neural_featurewise model from the training pipeline first.")
else:
    rows_mlp = []
    for project_name, info in projects.items():
        result = mlp_predict_project(mlp_checkpoint, info["features"])
        m = result["multipliers"]
        g = result["group_multipliers"]
        raw = result["raw_output"]
        rows_mlp.append({
            "project":       project_name,
            "clamped":       "⚠ yes" if result["any_clamped"] else "ok",
            # Raw network outputs (before clamping) — useful to see if model is saturating
            "raw_geo":       round(raw[0], 4),
            "raw_app":       round(raw[1], 4),
            "raw_den":       round(raw[2], 4),
            # Clamped group multipliers
            "geo_mult":      round(g["geometry_lr_mult"], 4),
            "app_mult":      round(g["appearance_lr_mult"], 4),
            "den_mult":      round(g["densification_mult"], 4),
            # 8 expanded individual multipliers
            "pos_lr":        round(m["position_lr_init_mult"], 4),
            "scl_lr":        round(m["scaling_lr_mult"], 4),
            "rot_lr":        round(m["rotation_lr_mult"], 4),
            "fea_lr":        round(m["feature_lr_mult"], 4),
            "opa_lr":        round(m["opacity_lr_mult"], 4),
            "lam_dssim":     round(m["lambda_dssim_mult"], 4),
            "den_grad":      round(m["densify_grad_threshold_mult"], 4),
            "opa_thr":       round(m["opacity_threshold_mult"], 4),
        })

    df_mlp = pd.DataFrame(rows_mlp)
    obj = mlp_checkpoint.get("objective", "?")
    print(f"=== MLP PREDICTION  (featurewise, objective={obj}) ===\n")
    print(df_mlp.to_string(index=False))

    clamped = df_mlp[df_mlp["clamped"] == "⚠ yes"]
    if not clamped.empty:
        print(f"\n⚠  {len(clamped)} project(s) had raw outputs outside bounds that were clamped to bounds.")
        print("   This means the MLP is saturating. More training data or tighter regularisation may help.")
```

## Section 3 — Side-by-side comparison

Compare ridge vs MLP group multipliers per project. **Run Section 1 and Section 2 first.**

```python
# ── COMPARISON: ridge vs MLP group multipliers ────────────────────────────────
ridge_ready = "df_ridge" in dir() and df_ridge is not None
mlp_ready   = "df_mlp"   in dir() and df_mlp   is not None

if not ridge_ready and not mlp_ready:
    print("Run Section 1 (ridge) and/or Section 2 (MLP) first.")
else:
    rows_cmp = []
    for project_name in projects:
        row = {"project": project_name}
        if ridge_ready:
            r = df_ridge[df_ridge["project"] == project_name]
            if not r.empty:
                row.update({"ridge_geo": float(r["geo_mult"].iloc[0]),
                             "ridge_app": float(r["app_mult"].iloc[0]),
                             "ridge_den": float(r["den_mult"].iloc[0]),
                             "ridge_sig": r["signal"].iloc[0]})
        if mlp_ready:
            r = df_mlp[df_mlp["project"] == project_name]
            if not r.empty:
                row.update({"mlp_geo": float(r["geo_mult"].iloc[0]),
                             "mlp_app": float(r["app_mult"].iloc[0]),
                             "mlp_den": float(r["den_mult"].iloc[0])})
        if ridge_ready and mlp_ready:
            row["Δ_geo"] = round(row.get("mlp_geo", 1.0) - row.get("ridge_geo", 1.0), 4)
            row["Δ_app"] = round(row.get("mlp_app", 1.0) - row.get("ridge_app", 1.0), 4)
            row["Δ_den"] = round(row.get("mlp_den", 1.0) - row.get("ridge_den", 1.0), 4)
        rows_cmp.append(row)

    df_cmp = pd.DataFrame(rows_cmp)
    print("=== RIDGE vs MLP GROUP MULTIPLIER COMPARISON ===\n")
    print(df_cmp.to_string(index=False))
```

```python
# ── COMPARISON: grouped bar chart ─────────────────────────────────────────────
import matplotlib.pyplot as plt

ridge_ready = "df_ridge" in dir() and df_ridge is not None
mlp_ready   = "df_mlp"   in dir() and df_mlp   is not None

if ridge_ready or mlp_ready:
    project_names = list(projects.keys())
    x = np.arange(len(project_names))
    w = 0.35

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle("Group multipliers per project — Ridge (blue) vs MLP (orange)", fontsize=12)

    for ax, (ridge_col, mlp_col, title, (lo, hi)) in zip(axes, [
        ("ridge_geo", "mlp_geo", "geometry_lr",   (0.5, 1.5)),
        ("ridge_app", "mlp_app", "appearance_lr", (0.5, 1.5)),
        ("ridge_den", "mlp_den", "densification", (0.7, 1.3)),
    ]):
        def _get(df, col, pname, fallback=1.0):
            r = df[df["project"] == pname]
            return float(r[col].iloc[0]) if (not r.empty and col in df.columns) else fallback

        if ridge_ready and ridge_col in df_cmp.columns:
            rv = [_get(df_cmp, ridge_col, p) for p in project_names]
            ax.bar(x - w/2, rv, w, label="Ridge", alpha=0.8)
        if mlp_ready and mlp_col in df_cmp.columns:
            mv = [_get(df_cmp, mlp_col, p) for p in project_names]
            ax.bar(x + w/2, mv, w, label="MLP", alpha=0.8)

        ax.axhline(1.0, color="gray", linestyle="--", linewidth=0.8, label="neutral")
        ax.set_title(title)
        ax.set_xticks(x)
        ax.set_xticklabels(project_names, rotation=60, ha="right", fontsize=7)
        ax.set_ylabel("multiplier")
        ax.set_ylim(lo - 0.05, hi + 0.05)
        ax.legend(fontsize=8)

    plt.tight_layout()
    plt.show()
```
