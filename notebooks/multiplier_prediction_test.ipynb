# Featurewise Quality Scoring Notebooks

The previous combined notebook has been split into fully independent notebooks:

- `three_model_featurewise_ridge_quality_scoring.ipynb` - earlier three-model Featurewise Ridge quality-scoring notebook.
- `three_model_featurewise_mlp_quality_scoring.ipynb` - earlier three-model Featurewise MLP quality-scoring notebook.
- `featurewise_ridge_quality_scoring.ipynb` - Featurewise Ridge quality-scoring notebook.
- `featurewise_mlp_quality_scoring.ipynb` - Featurewise MLP quality-scoring notebook.

Each notebook contains its own configuration, schema constants, data loading, training, prediction, candidate-score checks, and final effective parameter calculations. No backend package imports are used.
